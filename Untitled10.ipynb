```python
import csv
import urllib.request
from bs4 import BeautifulSoup


url='https://www.booking.com/reviews/au/hotel/west-sydney.en-gb.html'
req = urllib.request.Request(
    url,
    headers={
        'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36',
    }
)


f = urllib.request.urlopen(req)
soup = BeautifulSoup(f.read().decode('utf-8'),'html.parser')
reviews = soup.findAll("li", {"class": "review_item clearfix "})

review_list = []
for review in reviews:
    review_list.append(review.find("div", {"class": "review_item_review_content"}).text)

outputfilename = 'reviews_in_csv'
with open(outputfilename, 'w', newline='') as myfile:
     wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
     wr.writerow(review_list)
```
